# 1. Growth Of Function
Time Complexity/Order of Growth defines the amount of time taken by any program with respect to the size of the input. Time Complexity specifies how the program would behave as the order of size of input is increased. So, Time Complexity is just a function of size of its input. We will go through some of basic and most common time complexities such as:

- *Constant Time Complexity O(1) : constant running time*
- *Linear Time Complexity O(n) : linear running time*
- *Logarithmic Time Complexity O(log n) : logarithmic running time*
- *Log-Linear Time Complexity O(n log n) : log-Â­linear running time*
- *Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)*
- *Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)*

## Constant time Complexity:O(1)
An algorithm is said to have constant time with order O (1) when it is not dependent on the input size n. Irrespective of the input size n, the runtime will always be the same. See example below

```python
# Load the Python libraries
import matplotlib.pyplot as plt
```

```python
# Prob_1

def add_two_numbers(a, b):
    return a + b

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    add_two_numbers(n, n + 1)
    time_values.append(1)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(1) Constant Time Complexity for Addition")
plt.show()
```

```python
#Prob_2:Take another example and write the code that runs in fixed amount of time or has fixed number of steps of execution no matter what is the size of input has constant time complexity.
#similar to example-1, also plot the graph (eg. even_odd )
import time
import matplotlib.pyplot as plt
import random

# Constant Time Function: Check Even/Odd
def is_even(n):
    return n % 2 == 0

# Experiment to measure time taken
def experiment():
    sizes = [1000, 2000, 5000, 10000, 20000, 50000, 100000]  # input sizes
    times = []

    for n in sizes:
        numbers = [random.randint(1, 10**6) for _ in range(n)]  # generate n random numbers

        start_time = time.time()
        # Perform only ONE check (constant time)
        is_even(numbers[0])
        end_time = time.time()

        times.append(end_time - start_time)

    # Plotting the graph
    plt.plot(sizes, times, marker='o', linestyle='-', color='g')
    plt.xlabel("Number of elements (n)")
    plt.ylabel("Time taken (seconds)")
    plt.title("Constant Time Example (Even/Odd Check)")
    plt.grid(True)
    plt.show()

# Run the experiment
experiment()
```

#Linear Time Complexity O(n) : linear running time
- *An algorithm is said to have a linear time complexity when the running time increases linearly with the length of the input. When the function involves checking all the values in input data, with this order O(n).*

```python
def linear_time_sum(numbers):
    total = 0
    for value in numbers:
        total += value
    return total

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    data = list(range(n))
    linear_time_sum(data)
    time_values.append(n)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(n) Linear Time Complexity")
plt.show()
```

```python
#Prob_3:Take another example and write the code that runs in linear amount of time no matter what is the size of input..
#similar to example-2, also plot the graph (eg. even_oddcounts elements greater than zero)

import time
import matplotlib.pyplot as plt
import random

def count_positive(numbers):
    count = 0
    for value in numbers:   # This loop runs 'n' times, where 'n' is the length of 'numbers'
        if value > 0:
            count += 1
    return count

n_values = list(range(1, 10001, 500))  # Input sizes ranging from 1 to 10000
time_values = []

for n in n_values:
    data = [random.randint(-100, 100) for _ in range(n)] # Generate a list of 'n' random numbers

    start_time = time.time()
    count_positive(data)
    end_time = time.time()

    time_values.append(end_time - start_time)

plt.figure()
plt.plot(n_values, time_values, marker='o')
plt.xlabel("Input size (n)")
plt.ylabel("Time (seconds)")
plt.title("O(n) Linear Time Complexity: Count Positive Numbers")
plt.grid(True)
plt.show()
```

#Logarithmic Time Complexity O(log n) : logarithmic running time
- *An algorithm is said to have a logarithmic time complexity when it reduces the size of the input data in each step. This indicates that the number of operations is not the same as the input size. The number of operations gets reduced as the input size increases. Algorithms are found in binary tree or binary search functions. This involves the search of a given value in an array by splitting the array into two and starting searching in one split. This ensures the operation is not done on every element of the data.*

```python
#Prob_4 Take Suitable example and write function that runs in logarithic time. plot graph as well similar to above example.

import matplotlib.pyplot as plt
import time
import random

def binary_search(arr, target):
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1 # Target not found

n_values = list(range(100, 100001, 5000)) # Input sizes from 100 to 100,000
time_values = []

for n in n_values:
    # Generate a sorted list of 'n' numbers
    data = sorted([random.randint(0, n*2) for _ in range(n)])
    # Choose a target that is likely to be found (or near the end of search)
    target = data[random.randint(0, n-1)] if n > 0 else 0

    start_time = time.time()
    binary_search(data, target)
    end_time = time.time()

    time_values.append(end_time - start_time)

plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values, marker='o', linestyle='-', color='purple')
plt.xlabel("Input size (n)")
plt.ylabel("Time (seconds)")
plt.title("O(log n) Logarithmic Time Complexity: Binary Search")
plt.grid(True)
plt.show()
```

#*Log-Linear Time Complexity O(n log n) : log-Â­linear running time*


* When we call a Logarithmic Time Algorithm inside a loop, it would
result into a Log-Linear Time Complexity program.




```python
#Prob_5 Take Suitable example and write function that runs in log-linear time. plot the graph as well, similar to above example.

import matplotlib.pyplot as plt
import time
import random

def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]

    left_half = merge_sort(left_half)
    right_half = merge_sort(right_half)

    return merge(left_half, right_half)

def merge(left, right):
    merged_list = []
    left_idx, right_idx = 0, 0

    while left_idx < len(left) and right_idx < len(right):
        if left[left_idx] < right[right_idx]:
            merged_list.append(left[left_idx])
            left_idx += 1
        else:
            merged_list.append(right[right_idx])
            right_idx += 1

    merged_list.extend(left[left_idx:])
    merged_list.extend(right[right_idx:])
    return merged_list

n_values = list(range(1, 2001, 100))  # Input sizes from 1 to 2000
time_values = []

for n in n_values:
    data = [random.randint(0, 1000) for _ in range(n)] # Generate a list of 'n' random numbers

    start_time = time.time()
    merge_sort(data)
    end_time = time.time()

    time_values.append(end_time - start_time)

plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values, marker='o', linestyle='-', color='b')
plt.xlabel("Input size (n)")
plt.ylabel("Time (seconds)")
plt.title("O(n log n) Log-Linear Time Complexity: Merge Sort")
plt.grid(True)
plt.show()
```

```python
# Write a function to perform quick sort and show that it has Logarithmic Time Complexity O(log n)
import random
import matplotlib.pyplot as plt
import math

max_depth = 0

def quick_sort_depth(arr, depth=1):
    global max_depth
    if depth > max_depth:
        max_depth = depth
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort_depth(left, depth + 1) + middle + quick_sort_depth(right, depth + 1)

input_sizes = list(range(1, 102, 5))  # from 1 to 101, step 5
depths = []

for n in input_sizes:
    max_depth = 0
    arr = random.sample(range(n * 3), n)  # unique random elements
    quick_sort_depth(arr)
    depths.append(max_depth)

# Calculate theoretical log2(n)
log_n = [math.log2(x) if x > 0 else 0 for x in input_sizes]

plt.figure(figsize=(10,6))
plt.plot(input_sizes, depths, label='Measured Recursion Depth', marker='o')
plt.plot(input_sizes, log_n, label='logâ(n) (Theoretical)', linestyle='--')
plt.xlabel('Input Size (n)')
plt.ylabel('Recursion Depth')
plt.title('Quick Sort Recursion Depth vs Input Size')
plt.legend()
plt.grid(True)
plt.show()

```
